```python
# Calculation function
def performCalculation(operation):
    
    if operation == '+':
        op_name = 'addition'
    elif operation == '-':
        op_name = 'subtraction'
    elif operation == '*':
        op_name = 'multiplication'
    else: #operation is '/'
        op_name = 'division'
    
    print('\n ***************')
    print('Please input the two numbers to perform your ' + op_name +'.')
    print('Input the numbers one at a time.')
    print("Type 'stop' to end the portion of the program without performing any calculations. \n")
    while True:
        first_num = input("What is the first number? ")
        if str(first_num.lower()) == 'stop': 
            print('You have chosen to exit this portion of the program. \n')
            print('\n ***************')
            return
        elif len(first_num) == 0:
            print("\n We did not detect a number. \n"
                 "Can you please try again? \n")
        else:
            first_num = float(first_num)
            break #have the first number        

    #print('Input the the second number. \n')
    while True:
        second_num = input("What is the second number? \n"
                                "Again, type 'stop' to end without performing any calculations. ")
        if str(second_num.lower()) == 'stop': 
            print('You have chosen to exit this portion of the program. \n')
            print('\n ***************')
            return
        elif len(second_num) == 0:
            print("\n We did not detect a number. \n"
                 "Can you please try again? \n")
        else:
            second_num = float(second_num)
            break #have the second number
    
    if operation == '+':
        result = first_num + second_num
    elif operation == '-':
        result = first_num - second_num
    elif operation == '*':
        result = first_num * second_num
    else: #operation is '/'
        result = first_num / second_num
    
    print("The result of the " + op_name + " is: " + str(result))
    print('\n ***************')
```

```python
# Average function
def calculateAverage():
    
    print('\n ***************')
    print("Let's start by telling us how many numbers you want to average.")
    print("Type 'stop' to end this portion of the program without performing any calculations. \n")
    while True:
        
        count_nums = input("How many numbers do you wish to input? ")
        
        if count_nums.lower() == 'stop': 
            print('You have chosen to exit this portion of the program. \n')
            print('\n ***************')
            return
        elif len(count_nums) == 0:
            print("\n We did not detect a number. \n"
                 "Can you please try again? \n")
        else:
            try:
                int(count_nums)
                if float(count_nums) <= 0:
                    print("\n You have input either zero or a negative number.")
                    print(" Please try again. \n")
                else:
                    count_nums = int(count_nums)
                    break #have the first number
            except ValueError:
                print('You have entered either a letter, a float number, or a non-integer character. \n')
                print('Please try again')            
    
    print(count_nums)
    total = 0.0
    average = 0.0
    nums_entered = 0
    for num in range(count_nums):
        print("So far you have entered " + str(num) + " of " + str(count_nums) + " values.")
        if num == 0:
            input_num = input("Please enter your first number. ")
        else:
            input_num = input("Please enter your next number. ")
        try:
            float(input_num)
            total = total + float(input_num)
            nums_entered = nums_entered + 1
        except ValueError:
            if input_num.lower() == 'stop': 
                print('You have chosen to exit this portion of the program. \n')
                print('\n ***************')
                return
            else:
                print('You have entered either a letter, or a non-integer character. \n')
                print('This value will not be calculated in the average.')
        
        print('total is ' + str(total))
    
    average = total / nums_entered
    print('You said that you would enter ' + str(count_nums) + ' numbers.')
    print('You entered ' + str(nums_entered) + ' numbers.')
    print('The average of your entered numbers is ' + str(average) + '.')
    print('\n ***************')
```

```python
# this is the main program

possibleOps = ['+', '-', '*', '/', 'avg']
mathOps = ['+', '-', '*', '/']
print('This program will ask you what operation to perform, then take you there to perform.')
print('There are 5 possible operations: \n'
            '+ (for addition) \n'
              '- (for subtraction) \n'
              '* (for multiplication) \n'
              '/ (for division) \n'
              'avg (to calculate an average) \n')
print("Also, you can type 'stop' to end the full program. \n")
while True:
    perform_op = input("What operation would you like to perform? ") 
      
    if str(perform_op.lower()) == 'stop': 
        break
    elif perform_op.lower() not in possibleOps:
        print("\n We did not detect a correct operation to perform. \n"
                 "Can you please try again? \n")
    else:
        if perform_op in mathOps:
            #print('math')
            performCalculation(perform_op)
            #do the math
        else:
            #print('avg')
            calculateAverage()
            #do the average
        
print('\n ***************')      
print('The program is complete. Thank you and have a good day.')
```
