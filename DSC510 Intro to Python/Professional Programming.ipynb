```python
# add word function. 
# Adds each new word to the dictionary
# and increments the counter for each word
# no return value
def add_word(word_to_add, dict_name):
    
    if word_to_add not in dict_name:
        dict_name[word_to_add] = 1
    else:
        dict_name[word_to_add] += 1
    
```

```python
# process line function
# strips off punctuation characters, splits out words,
# makes all the letters lower case
# calls to add_word for every word in the line,
# once the processing is complete
# no return value
def Process_line(proc_line, input_dict):
    
    import string
    
    #strip out non-letter characters
    proc_line = proc_line.rstrip()
    proc_line = proc_line.translate(proc_line.maketrans('','', string.punctuation))
    
    #make all small letters
    proc_line = proc_line.lower()
    
    #split the words out from the line
    line_words = proc_line.split()
    
    for word in line_words:
        add_word(word, input_dict)
    
```

```python
# printing function
# receives a dictionary as input
# the dictionary has a list of words and frequencies of each word
# sorts the dictionary in order of word frequency
# prints the dictionary to the screen
def Pretty_print(print_dict):
    
    word_list = list()
    total_words = 0
    
    for key, value in list(print_dict.items()):
        word_list.append((value, key))
        
    word_list.sort(reverse=True)
    
    str1 = "The word '{}' appears {} time."
    str2 = "The word '{}' appears {} times."
    for key, value in word_list:
        if key>1:
            print(str2.format(value, key))
        else:
            print(str1.format(value, key))
        total_words = total_words + key
            
    print('\nThe total number of words in the file is ' + str(total_words) + '.')    
              
```

```python
# opens a file
# returns either the file's contents, 
# or the value 'null', if the file doesn't exist
def file_open(file_name):
    import os

    #set the directory to current
    #set an absolute path
    curDir = os.path.abspath('.') #sets the current directory to the one holding the file being executed.
    os.chdir(curDir) #sets the new current working directory
    cwd = os.getcwd()  #put the current working directory into string variable
    #print(cwd)
    
    #check to see if the file exists and assign to filepath variable
    if len(file_name)>0:
        #print(len(file_name))
        if os.path.isfile(file_name):
            #print('here1')
            filepath = (file_name)
            #print(filepath)
        else:
            #error, the file doesn't exist
            return 'null'

    #open file for full path and filename and strip the EOL from each string
    with open(filepath, 'r') as f:
        for line in f:
            line = line.rstrip("\n")
            #print(line)
        f.seek(0)
        content = f.readlines()
        return content
```

```python
# this is the main program

def main():
    
    import tkinter as tk
    from tkinter import filedialog
    
    #initialize the dictionary to process
    #we will pass this dictionary to the various functions
    count_dict = dict()

    root = tk.Tk()
    #To show only the dialog without any other GUI elements, 
    #you have to hide the root window using the withdraw method
    #got this from Stack Overflow
    root.withdraw()

    #prompt the user for the text file to process
    file_path = filedialog.askopenfilename()
    
    #open the file and put the contents into a variable
    file_text = file_open(file_path)
    
    if file_text == 'null':
        print('File does not exist, ending program.')
    else:  #good file to process
        for line in file_text:
            Process_line(line, count_dict)

        Pretty_print(count_dict)
    
    
```

```python
# this runs the program
if __name__ == "__main__":
    main()
```
