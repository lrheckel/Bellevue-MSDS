```python
# go to weather site function
def getWeather():
    
    import json
    import requests
    import pprint

    zip_code = ''
    city = ''
    weather_URL = ''
    good_data = True
    
    #these should go into a config file in a production program
    #and code here to read them
    weather_current = 'http://api.openweathermap.org/data/2.5/weather?'
    weather_forecast = 'http://api.openweathermap.org/data/2.5/forecast?'
    weather_key = 'APPID=3cf986326f263cf9d4896e04abee9120'

    #need to validate the input
    input_value = str(input("\n******************"
                            "\nPlease type a world-wide city name or US zip code, or type 'n' to exit. "
                            "\n   For cities whose names may exist in multiple countries:\n"
                            "           please add a comma and the two-letter country code."
                            "\n   If you are unsure of the country code, copy and paste\n"
                            "           this link into the address of your browser:\n"
                            "           https://en.wikipedia.org/wiki/ISO_3166-1_alpha-2#Officially_assigned_code_elements"
                            "\n   For US cities whose names may exist in multiple states:\n"
                            "           please add a comma, the two-letter state code, another comma, and the two-letter country code (US)"
                            "\n\n   Note that only US zip codes are currently supported."
                            "\n\nExamples:  44512  or  Manchester,UK  or  Manchester,NH,US   "))

    #check the input value for either zip or city
    if len(input_value)==5 and int(input_value):
        zip_code = input_value
        weather_URL_current=weather_current + "zip=" + zip_code + "&" + weather_key + "&units=imperial"
        weather_URL_forecast=weather_forecast + "zip=" + zip_code + "&" + weather_key + "&units=imperial"
        #print(weather_URL)
    else:
        city = input_value
        weather_URL_current=weather_current + "q=" + city + "&" + weather_key + "&units=imperial"
        weather_URL_forecast=weather_forecast + "q=" + city + "&" + weather_key + "&units=imperial"
        #print(weather_URL)
    
    #go to the URL to get the weather data
    try:
        print('\nAttempting to contact the weather site..................\n')
        response_current = requests.get(weather_URL_current)
        response_forecast = requests.get(weather_URL_forecast)
        
        response_json_curr = json.loads(response_current.text)
        response_json_fore = json.loads(response_forecast.text)
        
        
        if response_json_curr['cod'] == '404':
            return_key = print(str(input('The city/zip you entered could not be found\n'
                 'Please press the return key to try again.\n')))
            good_data = False
        
        else:  #good weather value, let's display the results
            print('Successfully contacted the weather site............\n')
            good_data = True
            #printWeather(response_json)
            
    except:   #only comes here if not able to connect to the URL
        print(response_current)
        return_key = print(str(input('There was an error in connecting to the weather site.\n'
                 'No weather report was returned.\n'
                 'Please press the return key to try again.\n')))
        good_data = False
        
    if good_data:
        printCurrentWeather(response_json_curr)
        printForecastWeather(response_json_fore)
```

```python
def printForecastWeather(forecast_data):
    
    import json
    import pprint
    import datetime

    # set the variables for the forecast
    f_city = forecast_data['city']
    f_list = forecast_data['list']

    
    #put the min and max temps into lists to find daily min and max
    #openweather returns 38 values over 5 days, spaced out at 3 hour intervals
    #Day 1 = 0-7
    #Day 2 = 8-15
    #Day 3 = 16-23
    #Day 4 = 24-31
    #Day 5 = 32-37
    
    temp_min = [None] * 38
    temp_max = [None] * 38
    for i in range(38):
        temp_min[i]=(f_list[i]['main']['temp_min'])
        temp_max[i]=(f_list[i]['main']['temp_max'])
        #print(str(i) + ',' + str(temp_min[i]) + "," + str(temp_max[i]))
    
    
    # output the 5 day forecast
    #This is the header printout
    print('\n***********')
    print('And here is the forecast for the next 5 days for %s:' %f_city['name'] + '\n')
    
    
    #loop through the data to display the days
    #I picked values that represent the middle of the day,
    #for wind speed/direction and humidity,
    #and those middle of the day ones are the values of "i".
    for i in (5, 13, 21, 28, 35):
        #*******************
        # set the variables
        #foreMinTemp = round(f_list[i]['main']['temp_min'])
        #foreMaxTemp = round(f_list[i]['main']['temp_max'])
        
        if 0 <= i <= 7:
            foreMinTemp = round(min(temp_min[0:7]))
            foreMaxTemp = round(max(temp_max[0:7]))
        elif 8 <= i <= 15:
            foreMinTemp = round(min(temp_min[8:15]))
            foreMaxTemp = round(max(temp_max[8:15]))
        elif 16 <= i <= 23:
            foreMinTemp = round(min(temp_min[16:23]))
            foreMaxTemp = round(max(temp_max[16:23]))
        elif 24 <= i <= 31:
            foreMinTemp = round(min(temp_min[24:31]))
            foreMaxTemp = round(max(temp_max[24:31]))
        else:
            foreMinTemp = round(min(temp_min[32:37]))
            foreMaxTemp = round(max(temp_max[32:37]))
        
        foreHumid = round(f_list[i]['main']['humidity'])
        
        if 45 < f_list[i]['wind']['deg'] <= 135:
            wind_dir = 'East'
        elif 135 < f_list[i]['wind']['deg'] <= 225:
            wind_dir = 'South'
        elif 225 < f_list[i]['wind']['deg'] <= 295:
            wind_dir = 'West'
        else:
            wind_dir = 'North'

        wind_speed = str(round(f_list[i]['wind']['speed']))
        
        #variables are all set, now go ahead and print
        print("The weather forecast for " + 
          str(datetime.datetime.fromtimestamp(int(f_list[i]['dt'])).strftime('%B %d, %Y')) + ':\n')
        
        print("Overall, the day will be " + str(f_list[i]['weather'][0]['main']).lower() + ' with '
              + str(f_list[i]['weather'][0]['description']).lower() + '.')

        print("The day's high temperature will be " + str(foreMaxTemp) + ' degrees Fahrenheit.')
        print("The day's low  temperature will be " + str(foreMinTemp) + ' degrees Fahrenheit.')
        print("The day's humidity will be " + str(foreHumid) + '%.' )
        print("The wind will be from the " + wind_dir + ' at ' + wind_speed + ' mph.')
        print('\n****************************\n')
    
```

```python
# this is the main program function
def main():
    
    counter = 0
    
    print('This program will ask you if you would like to see weather for a location.')
    print('Each time you answer positively, it will ask you for a location, either zip or city, and then return the weather.')
    print("When you no longer wish to see any more weather, you can exit the program.")
    print('***************\n')
    while True:
        if counter == 0:
            perform_op = input("Would you like to get a weather report? \n"
                          "Type 'y' or 'Y' for a weather report, or type 'n' or 'N' to exit. ")
        else:
            perform_op = input("Would you like to get another weather report? \n"
                          "Type 'y' or 'Y' or an additional report, or type 'n' or 'N' to exit. ")

        if str(perform_op.lower()) == 'n': 
            break
        elif str(perform_op.lower()) == 'y': 
            #print('yes')
            getWeather()
            counter += 1
        else:
            print("\nThe program did not detect a 'Y' or 'N' answer. \n"
                     "Can you please answer again?")
        
    print('\n***************')
    print('The program is complete. Thank you and have a good day.')
```

```python
# this runs the program
if __name__ == "__main__":
    main()
```
