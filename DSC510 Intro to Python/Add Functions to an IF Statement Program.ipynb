```python
#Input for the company name
def get_company_name(prompt):
    while True:
        input_value = str(input("What is the name of your company? \n"
                                "Type 'stop' to end without purchasing any cable. "))
        if input_value.lower() == 'stop': 
            return input_value.lower()
        elif len(input_value) == 0:
            print("\n We did not detect your company name. \n"
                 "Can you please try again? \n")
        else:
            print("Welcome, " + str(input_value))
            return input_value
        
#Input for number feet to buy
def get_number_feet(prompt):
    while True:
        try:
            feet_value = float(input(prompt))
        except ValueError:
            print()
            print("What you entered is not a number. Please try again.")
            continue

        if feet_value < 0:
            print()
            print("The length of cable must be greater than zero, if you wish to purchase.")
            print(" Please try again. ")
            print()
            continue
        else:
            return feet_value
        
#compute the price that the user will pay
def compute_cost(feet, price_per_foot):
    cost = price_per_foot * feet
    cost = '{0:.2f}'.format(cost)
    return cost

#-------------------------------
#Program Execution starts here
#-------------------------------

#Initialize variables    
number_feet = 0
cable_cost = 0.0
price_per_foot = 0.0

#Welcome message
print("Welcome to the Fiber Optic Cable Company.")
print("Our aim is to satisy your cable needs.")
print("Let's start by asking for some basic information. \n")

company_name = get_company_name("What is the name of your company?")

#blank line for readability
print()

if company_name != 'stop':
    number_feet = get_number_feet("How many feet of cable would you like to purchase? \n"
                                      "Input the length in feet and tenths of feet. \n"
                                      "Input 0 if you do not wish to purchase any cable. " )
    #get the price per foot, only need if user wants to continue    
    if number_feet > 500:
        price_per_foot = .50
    elif number_feet > 250:
        price_per_foot = .70
    elif number_feet > 100:
        price_per_foot = .80
    else:
        price_per_foot = .87
    
    
if number_feet > 0:
    cable_cost = compute_cost(number_feet, price_per_foot)
    
#blank line for readability
print()

if number_feet > 0: #provided company name and number of feet to purchase
    print(company_name + ', you have chosen to purchase ' + str(number_feet) + ' feet of cable.')
    print('Your cost for this is $' + str(cable_cost))
    print('Please contact us directly to place your order.')
elif company_name != 'stop': #provided company name, and 0 for feet of cable 
    print(company_name + ", you have chosen not to purchase any cable today." + "\n")
    print("Thank you. Have a nice day.")
else:  #did not provide company name
    print('You have chosen not to give your company name or purchase cable today.')
    print("Thank you. Have a nice day.")


```
